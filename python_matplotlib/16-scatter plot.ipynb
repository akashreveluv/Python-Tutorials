# Scatter plots 
are widely used graphs, especially they come in handy in visualizing a problem of regression. In the following example, we feed in arbitrarily created data of height and weight and plot them against each other. We used xlim() and ylim() methods to set the limits of X-axis and Y-axis respectively.

```python
import matplotlib.pyplot as plt
import numpy as np

heights = [169, 142, 170, 149, 150, 180, 170, 155, 166, 174]
weights = [80, 69, 79, 72, 70, 88, 78, 77, 80, 75]

plt.xlim(140,200)
plt.ylim(60,100)
plt.scatter(heights, weights)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Scatterplot')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

heights = [169, 142, 170, 149, 150, 180, 170, 155, 166, 174]
weights = [80, 69, 79, 72, 70, 88, 78, 77, 80, 75]

ax = plt.axes(projection="3d")
ax.scatter3D(heights, weights)
ax.set_xlabel('Height')
ax.set_ylabel('Weight')
plt.title('Scatterplot')
plt.grid()
plt.show()
```
