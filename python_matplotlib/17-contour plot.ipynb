```python
import matplotlib.pyplot as plt
import numpy as np

width_of_panel = 4
height_of_panel = 3
d = 1000
plt.figure(figsize=(width_of_panel, height_of_panel), dpi=d)
x = np.arange(0, 25, 1)
y = np.arange(0, 25, 1)
x, y = np.meshgrid(x, y)
z = np.sin(x * np.pi / 2) + np.cos(y * np.pi / 3) # np.sin(x * np.pi / 2) + np.cos(y * np.pi / 3)

plt.contour(x, y, z)
```
