# Matplotlib API provides the bar() function 
that can be used in the MATLAB style use as well as object oriented API. The signature of bar() function to be used with axes object is as follows âˆ’

ax.bar(x, height, width, bottom, align)

```python
import matplotlib.pyplot as plt

langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]

plt.bar(langs,students,color='green')
plt.show()
```

To make horizontal bar graphs use method barh() Also we can pass an argument (with its value)xerr or yerr (in case of the above vertical bar graphs) to depict the variance in our data as follows:

```python
import matplotlib.pyplot as plt

langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
variance = [5,4,7,6,3]

plt.barh(langs,students,xerr=variance,color='green')
plt.show()
```

To create horizontally stacked bar graphs we use the bar() method twice and pass the arguments where we mention the index and width of our bar graphs in order to horizontally stack them together. Also, notice the use of two other methods legend() which is used to show the legend of the graph and xticks() to label our x-axis based on the position of our bars.

```python
import matplotlib.pyplot as plt
import numpy as np

langs = ['C', 'C++', 'Java', 'Python', 'PHP']
boys_marks = [23,17,35,29,12]
girls_marks = [21,40,32,18,30]

index = np.arange(5)
width = 0.30

plt.bar(index, boys_marks, width, color='blue', label='Boys Marks')
plt.bar(index+width, girls_marks, width, color='green', label='Girls Marks')
plt.title("Students Result")

plt.xlabel("Languages")
plt.ylabel("Marks")
plt.xticks(index+width/2, langs)
plt.legend(loc='best')

plt.show()
```

Similarly, to vertically stack the bar graphs together, we can use an argument bottom and mention the bar graph which we want to stack below as its value.

```python
import matplotlib.pyplot as plt
import numpy as np

langs = ['C', 'C++', 'Java', 'Python', 'PHP']
boys_marks = [23,17,35,29,12]
girls_marks = [21,40,32,18,30]

index = np.arange(5)
width = 0.30

plt.bar(index, boys_marks, width, color='blue', label='Boys Marks')
plt.bar(index, girls_marks, width, color='green', label='Girls Marks', bottom=boys_marks)
plt.title("Students Result")

plt.xlabel("Languages")
plt.ylabel("Marks")
plt.xticks(index, langs)
plt.legend(loc='best')

plt.show()
```
