```python
# DICTIONARY

student = {"Name": "Akash",
           "Roll No": 20,
           "Course": {"Sub1": "DBMS", "Sub2": "OS", "Sub3": "AI"},
           "Class": "4th Sem"}
print(student)
print(student["Class"], student["Course"])  # We pass key- Name,Roll No,Course,Class and get value.
student["Name"] = "Reveluv"  # Edit in dict.
student["Roll No"] = "30"
print(student)
del student["Course"]  # Delete from dict.
print(student)
student.update({"Age": 26, "Sex": "Male"})  # Update or Add new value in dict.
print(student)
std = student.copy()  # Copy the dict Student as it is. If not use copy(), then any change will effect real dict.
print(std)
del std["Class"]  # Value of std (copy of student) deleted, but not delete in Student.
print(std)
print(student)
print(student.keys())  # To show all key values of dict.
print(student.items())  # To show all item values of dict.
student = {"Name": "Akash",
           "Roll No": 20,
           "Course": {"Sub1": "DBMS", "Sub2": "OS", "Sub3": "AI"},
           "Class": "4th Sem"}
print(student["Course"])  # Dict of dict.
print(student["Course"]["Sub2"])  # Particular value shows in dict of dict.
del student["Course"]["Sub2"]  # Delete from dict of dict
print(student["Course"])
student["Course"].update({"Sub4": "PHP", "Sub5": "MySQL"})
print(student)

# -------------------------1st scenario---------------------------------------
# COVID-19 Dict.
# Data at the time of IST 6 P.M, 27-05-2020 from Worldometer
covid = {
    "Total Cases": {"India": 153247, "South Korea": 11265, "Japan": 16623, "USA": 1725881, "Russia": 370680},
    "New Cases": {"India": "+2,454", "South Korea": "+40", "Japan": "+42", "USA": "+606", "Russia": "+8,338"},
    "Total Death": {"India": 4365, "South Korea": 269, "Japan": 846, "USA": 100625, "Russia": 3968},
    "New Death": {"India": "+21", "South Korea": "+2", "Japan": "+16", "USA": "+53", "Russia": "+161"}
}
country = input("Enter your country name:")
search = input("Total Cases\nNew Cases\nTotal Death\nNew Death\nWhich option you want to select ? :")
print(covid[search][country])
# country name input save in country variable and option save in search variable.
# we made dict. with options name 1st, and then under option name we put country details as dict of dict.
# thats why in print(covid[search][country]), we 1st mention search and then mention country.

# -------------------------2nd scenario---------------------------------------
covid = {
    "India": {"Total Cases": 153247, "New Cases": "+2,454", "Total Death": 4365, "New Death": "+21"},
    "South Korea": {"Total Cases": 11265, "New Cases": "+40", "Total Death": 269, "New Death": "+2"},
    "Japan": {"Total Cases": 16623, "New Cases": "+42", "Total Death": 846, "New Death": "+16"},
    "USA": {"Total Cases": 1725881, "New Cases": "+606", "Total Death": 100625, "New Death": "+53"},
    "Russia": {"Total Cases": 370680, "New Cases": "+8,338", "Total Death": 3968, "New Death": "+161"}
}
country = input("Enter your country name:")
search = input("Total Cases\nNew Cases\nTotal Death\nNew Death\nWhich option you want to select ? :")
print(covid[country][search])
```
