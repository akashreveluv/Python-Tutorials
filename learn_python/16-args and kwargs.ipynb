```python
def function_name(a, b, c):
    print(a, b, c)


function_name("akash", "yoona", "yeri")

# # *args------------------------------------------------------------------------------------
#


def function_name(*args):  # 1st use normal arguments then *args, then *kwargs
    # print(s)
    for item in args:
        print(item)


lst = ["akash", "yoona", "yeri", "sohyun"]
tpl = ("akash", "yoona", "yeri", "sohyun")
str = "I am a ReveLuv"

function_name(*lst)
function_name(*tpl)
function_name(*str)  # put * before tpl or lst to avoid () or [] in output respectively.
# if we put * before str we get all letters in str in new line
# #
# # # **kwargs----------------------------------------------------------------------------------
# #
# #
def function_name(**kwargs):  # 1st use normal arguments then *args, then *kwargs
    for key, value in kwargs.items():
        print(key, value)


dict = {"f1": "akash", "f2": "yoona", "f3": "yeri", "f4": "sohyun"}
function_name(**dict)
#
#
# # *args **kwargs--------------------------------------------------------------------------
#
def function_name(*args, **kwargs):  # 1st use normal arguments then *args, then *kwargs
    for item in args:
        print(item)
    for key, value in kwargs.items():
        print(key, value)


lst = ["akash", "yoona", "yeri", "sohyun"]
tpl = ("akash", "yoona", "yeri", "sohyun")
str = "I am a ReveLuv"
dict = {"f1": "akash", "f2": "yoona", "f3": "yeri", "f4": "sohyun"}

function_name(*lst, *tpl, **dict)

# --------------------------------------------------------------------------------------------
```
