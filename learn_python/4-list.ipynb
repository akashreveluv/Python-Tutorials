```python
# LIST

friends = ["Sohyun", "Yerim", "Joy", "Eunha", "Natty"]
numbers = [22, 77, 4, 9, 10, 34, 100, 53]
# if use quotation with 714,
# then being string, otherwise integer.
print(friends)
print(friends[2])  # print index 0 of the list, 0 is sohyun, 1 is yerim, 2 is joy.... so on.
print(friends[-2])  # -2 from 2nd last
print(friends[0:4])  # 0:4, 0-1-2-3, 1 less from 4th position.
print(friends[-3:-1])  # -1, 1 less position, so till -2nd show.
print(numbers)
print(numbers[0:5])
print(numbers[0:5:2])  # :2 means showing list from 0th to 4th with skip 1 number.
print(numbers[:5:2])
print(numbers[::2])
print(numbers[::2000])
# only print 22, because there no numbers of value till 2000,
# so can't show value after skip 1999.
print(numbers[::-2])  # -2 skip 1 number from reverse and show reverse result.
print(numbers[0:7:-2])  # show [] nothing, problem with ::-value
print(numbers.sort())  # o/p is none
# Sol:
numbers.sort()  # ascending order
print(numbers)
numbers.reverse()  # descending order
print(numbers)
#  We can sort a list only if list made with all strings or all numbers.
friends.sort()  # ascending order of 1st letter.
print(friends)
friends.reverse()  # descending order of 1st letter.
print(friends)
numbers.append(89)
print(numbers)
friends.append("Tzuyu")  # add new value in list at end.
print(friends)
friends.insert(3, "Yoona")  # if we want add value in middle of list, 1st mention index no. then new value.
print(friends)
numbers.insert(3, 14)
print(numbers)
numbers.remove(34)  # we pass actual value in remove(34), not a index no.
print(numbers)
numbers.pop()  # pop remove last element
print(numbers)
numbers.pop(4)  # we pass index no. in pop(4,5), not a value.
print(numbers)
friends[3] = "Lisa"  # we can edit value in this way , friends[3] is index no. to change with Lisa
print(friends)
numbers[4] = 61
print(numbers)
print(len(friends))  # length of index no. in friends list
print(len(numbers))  # length of index no. in numbers list
print(max(numbers))  # show max value
print(min(numbers))  # show min value

#Merge 2 lists
new_list = [friends, numbers]
print(new_list)  # 0th index- [----friends----], 1th index- [-----numbers------],.......
print(new_list[1][2])  # 2nd index of 0th/1st/... index


# -----------------------------Tuple--We can do everything as List, but can't modify tuple, means unmutable------------
tpl = (22, 77, 4, 9, 10, 34, 100, 53)
print(tpl)
print(tpl[3])
tpl1 = (22,)  # (22) only one value don't show as tuple, so we should (22,) to show as tuple
print(tpl1)
```
