```python
# FOR LOOP

for n in range(10):
    # loop use for show a line 10 times without written 10 nos of line.
    # we can also write (0, 9) in place of (10)
    print("Hello World!")

for n in range(10):  # we can also write (0, 9) in place of (10)
    print(n)  # print the values of n from 0 to 9.

for n in range(5, 10):  # starting from 5, end at 9.
    print(n)

# print even numbers between 0 to 100
for i in range(0, 10):
    if i % 2 == 0:
        print(i)
    else:
        print("Odd Present Here!")

mygf = ["tzuyu", "yerim", "sohyun", "yoona"]
for item in mygf:
    print("You are my dear :", item)

mygfdict = {"tzuyu": "Dancer", "yerim": "Singer", "sohyun": "Actress", "yoona": "Actress"}
# for item in mygfdict:  # only show keys, not value
#     print("You are my dear :", item)
# for item in mygfdict.values():  # show values of the keys
#     print("You are my dear :", item)
for name, designation in mygfdict.items():  # show both keys & values using items().
    print(name, "is our", designation)

mystring = "My name is Akash Reveluv"
vowels = "aeiouAEIOU"  # declare vowels in string
count = 0  # declare count's initial value 0
# for item in mystring:  # show each letter in new line including space.
#     print(item)
for item in mystring:
    if item in vowels:  # checking for vowels
        count = count + 1  # if find any vowel it continue count +1
print("Total no. of vowels in string :", count)
```
