```python
# EXAM AASAAN HAI RPS GAME SOLUTION with 5 TURN---------------------------------------------->

import random
choice = ["Rock", "Paper", "Scissor"]
while True:
    print("Rock Paper Scissor Game: ")
    you, comp = 0, 0
    for i in range(1, 6):
        print("Round", i, "Lets Start: ")
        print("Please select your choice: ")
        print("1. Rock", "2. Paper", "3. Scissor", sep="\n")
        yourchoice = int(input())
        if yourchoice == 1:
            print("You select Rock")
            yourchoice = "Rock"
        elif yourchoice == 2:
            print("You select Paper")
            yourchoice = "Paper"
        elif yourchoice == 3:
            print("You select Scissor")
            yourchoice = "Scissor"
        else:
            print("Invalid Choice")
            break
        compchoice = random.choice(choice)
        print("Computer select:", compchoice)

        if yourchoice == compchoice:
            you += 1
            comp += 1
            print("This round is drawn")
        elif (yourchoice == "Paper" and compchoice == "Rock") or (yourchoice == "Rock" and compchoice == "Scissor") or (yourchoice == "Scissor" and compchoice == "Paper"):
            you += 1
            print("You win this round")
        else:
            comp += 1
            print("Computer win this round")

        if you > comp:
            print("You win the Game: ")
            print("Score is:", "Your score:", you, "Computer score:", comp, sep=" ")
        elif comp > you:
            print("Computer win the Game: ")
            print("Score is:", "Computer score:", comp, "Your score:", you, sep=" ")
        else:
            print("Match Drawn: ")
            print("Score is:", "Your score:", you, "Computer score:", comp, sep=" ")

    game = input("Want to play again (yes/no): ")
    if game == 'yes' or game == 'Yes' or game == 'YES':
        continue
    else:
        break
```
